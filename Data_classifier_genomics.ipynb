```python
for i in df.columns :   # for accessing columns
  if ('Non_' in i and 'sequence' in i): # filter to get non-cds columns only
    unique = df[i].unique()            # here unique is a array containing unique sequences
    unique_values.append(int(len(unique)))  # to slice the final data into seperate tables
    

    for j in unique:  # iterating on unique sequences
      acession = [None]*df.shape[0]   # array of size= len(dataframe) i.e,no. of accession no.s present with all null values
      l=0
      for k in range(0,df.shape[0]): # iterating on rows or accession no.s
        if (j == df[i][k] ) :
          acession[l]= (df['Accession_no.'][k])     # to get all the accession no.s which have the particular ncds part
          l +=1 
      classification_data.insert(len(classification_data.columns), j , acession, True)
      acession.clear()
    
   
```

```python
def dataframe_creator(j):
  if j == 0 or j> len(unique_values):
    return;
  elif j==1:
    dummy_1 =0
  else :
    dummy_1 = sum(unique_values[:j-1])

  i = pd.DataFrame(classification_data.iloc[:,dummy_1:(dummy_1 + unique_values[j-1])])
  
  i.dropna(axis=0, how= 'all', inplace= True)
  i.dropna(axis=1, how = 'all',inplace = True)
  i.reset_index()
  return i;
```

```python
# for multisheets in a single xl sheet

writer = pd.ExcelWriter('/content/drive/MyDrive/Viral genomics /classification_data.xlsx') # change the path 

# choose a different file each time u run this part of code :)


for i in range(1,len(unique_values)+1):
  data_xl = dataframe_creator(i)
  print(i)
  display(data_xl)
  data_xl.to_excel(writer,sheet_name='ncds_{}'.format(i))
  data_xl.drop(data_xl.index,inplace=True)

writer.save()
```

```python
nan = df2.iloc[2,df2.shape[-1]-1] #to get nan 
new_df = pd.DataFrame(columns = ['Accession_no.','overall_ncds_length','genome_length','ncds_percentage'])

for j in range(df2.shape[0]):
  accession_no = df2['Accession_no.'][j]
  overall_ncds_len = 0
  for i in range(1,df2.shape[-1],5):
    if df2.iloc[j,i] == nan or type(df2.iloc[j,i])== float:
      continue
    length = int(df2.iloc[j,i].split()[-1])
    if int(df2.iloc[j,i].split()[0]) == 0 or int(df2.iloc[j,i].split()[0]) == -1 :
      ncds_len = 0 
    else :
      ncds_len = int(df2.iloc[j,i].split()[-1]) - int(df2.iloc[j,i].split()[0])
    overall_ncds_len += ncds_len
  lis = [{'Accession_no.':accession_no ,'overall_ncds_length':overall_ncds_len,'genome_length':length, 'ncds_percentage' : (overall_ncds_len/length)*100}]
  print(lis)
  new_df = new_df.append(lis, ignore_index = True)
```
